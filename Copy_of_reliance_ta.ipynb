```python
import pandas as pd
import numpy as np
import pandas_ta as ta
import matplotlib.pyplot as plt
from tvDatafeed import TvDatafeed, Interval
```

```python
tv= TvDatafeed(username=None, password=None)

data1 = tv.get_hist(
    symbol="RELIANCE",
    exchange="NSE",
    interval=Interval.in_daily,
    n_bars=100,
)

print(data1)

```

```python
type(data1)
```

```python
data1["SMA_20"] = ta.sma(data1.close, length=20)

data1["EMA_20"] = ta.ema(data1.close, length=20)

data1["RSI"] = ta.rsi(data1.close, length=14)

macd = ta.macd(data1.close)
data1 = pd.concat([data1, macd], axis=1)

data1.tail(20)
```

```python
data1.columns
```

```python
cols_to_float = ['MACD_12_26_9', 'MACDs_12_26_9', 'MACDh_12_26_9']
data1[cols_to_float] = data1[cols_to_float].astype(float)

plt.figure(figsize=(12,6))

plt.plot(data1.index,
         data1['MACD_12_26_9'],
         label='MACD',
         color='blue',
         linewidth=1.2)

plt.plot(data1.index,
         data1['MACDs_12_26_9'],
         label='Signal Line',
         color='orange',
         linewidth=1.2)

plt.bar(data1.index,
        data1['MACDh_12_26_9'],
        label='MACD Histogram',
        color='green',
        alpha=0.5)

plt.fill_between(data1.index,
                 data1['MACDs_12_26_9'],
                 data1['MACD_12_26_9'],
                 where=(data1['MACDh_12_26_9']>0),
                 color='green', alpha=0.25, label='Uptrend')

plt.fill_between(data1.index,
                 data1['MACD_12_26_9'],
                 data1['MACDs_12_26_9'],
                 where=(data1['MACDh_12_26_9']<0),
                 color='red', alpha=0.25, label='Downtrend')

plt.title("MACD indicator")
plt.xlabel("Date")
plt.ylabel("MACD Value")
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import pandas_ta as ta
supertrend = ta.supertrend(high=data1.high, low=data1.low, close=data1.close, length=7, multiplier=3)
data1 = pd.concat([data1, supertrend], axis=1)

data1.tail(20)
```

```python
d1 = data1.copy()
d1.index = pd.to_datetime(d1.index)

plt.figure(figsize=(15,10))

plt.plot(d1.index, d1['close'], label='Close', color='black', linewidth=1.2)

plt.plot(d1.index, d1['SUPERT_7_3.0'], label='Supertrend', color='blue', linewidth=2)

plt.fill_between(d1.index,
                 d1.low,
                 d1.high,
                 where=(d1['SUPERTd_7_3.0']==1),
                 color='green', alpha=0.1, label='Uptrend')

plt.fill_between(d1.index,
                 d1.low,
                 d1.high,
                 where=(d1['SUPERTd_7_3.0']==-1),
                 color='red', alpha=0.1, label='Downtrend')

plt.title('Supertrend Indicator')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```

```python
d2 = d1[['close', 'EMA_20', 'SMA_20', 'RSI']].copy()

d2['prevRSI'] = d2['RSI'].shift(1)

def rsi_decision(row):
  if row['prevRSI']<30 and row['RSI']>=30: return 'LONG'
  elif row['prevRSI']>70 and row['RSI']<=70: return 'SHORT'
  else: return 'HOLD'

d2['RSI_decision'] = d2.apply(rsi_decision, axis=1)

d2.head(50)
```

```python
d3 = d1[['close', 'high',
         'MACD_12_26_9', 'MACDh_12_26_9', 'MACDs_12_26_9',
         'SUPERT_7_3.0', 'SUPERTd_7_3.0', 'SUPERTl_7_3.0', 'SUPERTs_7_3.0']].copy()

d3['prev_macd'] = d3['MACD_12_26_9'].shift(1)
d3['prev_signal'] = d3['MACDs_12_26_9'].shift(1)

def macd_decision(row):
    if row['prev_macd'] < row['prev_signal'] and row['MACD_12_26_9'] >= row['MACDs_12_26_9']:
        return 'LONG'
    elif row['prev_macd'] > row['prev_signal'] and row['MACD_12_26_9'] <= row['MACDs_12_26_9']:
        return 'SHORT'
    else:
        return 'HOLD'

d3['MACD_decision'] = d3.apply(macd_decision, axis=1)

d3[['close', 'MACD_12_26_9', 'MACDs_12_26_9', 'MACD_decision']].tail(45)
```

```python
d3['prev_supert_dir'] = d3['SUPERTd_7_3.0'].shift(1)

def supert_decision(row):
    if row['prev_supert_dir'] == -1 and row['SUPERTd_7_3.0'] == 1:
        return 'LONG'
    elif row['prev_supert_dir'] == 1 and row['SUPERTd_7_3.0'] == -1:
        return 'SHORT'
    else:
        return 'HOLD'

d3['SUPERT_decision'] = d3.apply(supert_decision, axis=1)


d3[['close','SUPERT_7_3.0', 'SUPERTd_7_3.0', 'MACD_decision']].tail(45)
```
